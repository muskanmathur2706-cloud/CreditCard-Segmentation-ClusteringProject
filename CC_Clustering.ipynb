## **Credit** Card Customer Segmentation using K-Means Clustering
Aim - dividing bank customers in segments based on their spending behaviour

```python

import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv('CC GENERAL.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isnull().sum()#checking missing values
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)#filling missing values
```

```python
df.drop('CUST_ID', axis=1, inplace=True)#removing missing values
```

```python
scaler = StandardScaler()

scaled_data = scaler.fit_transform(df)#scaling since its important in clustering
```

```python
wcss = [] #wccs=within clusters sum of squares:the more a cluster is compact the lesser is wcss

for i in range(1,11):
    kmeans = KMeans(n_clusters=i,
                    random_state=42,
                    n_init=10)

    kmeans.fit(scaled_data)

    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8,5))
plt.plot(range(1,11), wcss, marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()
```

```python
kmeans = KMeans(
    n_clusters=4,
    random_state=42,
    n_init=10
)

clusters = kmeans.fit_predict(scaled_data)
```

create cluster column

```python
df['Cluster'] = clusters
```

```python
df.head()
```

# Cluster Distribution

```python
df['Cluster'].value_counts()
```

```python
df['Cluster'].value_counts().plot(
    kind='bar'
)

plt.title("Customer Segments")
plt.show()
```

(Visulization using PCA )--
PCA is used when there are so many features in the dataset,so to show them in 2D we use PCA.


```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

pca_data = pca.fit_transform(scaled_data)
```

```python
pca_df = pd.DataFrame(
    pca_data,
    columns=['PCA1','PCA2']
)

pca_df['Cluster'] = clusters
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    pca_df['PCA1'],
    pca_df['PCA2'],
    c=pca_df['Cluster']
)

plt.title("Customer Segmentation")
plt.xlabel("PCA1")
plt.ylabel("PCA2")

plt.show()
```
